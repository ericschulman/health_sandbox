```python
import pandas as pd
import numpy as np
```

```python
data_path = "../data/data_2019"
df = pd.read_csv(data_path + 'ACSDP5Y2019.DP05_data_with_overlays_2022-01-14T190209.csv')
```

```python
print('There are', df['GEO_ID'].nunique(), 'unique GeoIDs in the data.')
```

```python
df['County'] = df['GEO_ID'].str[-5:]
df['County'] = df['County'].astype(str)  
df['County'] = df['County'].str.lstrip('0') 
df
```

```python
data_path = "../data/"
df1 = pd.read_csv(data_path + 'Service_Area_PUF_2019.csv', encoding='cp1252')
df1 = df1.dropna(subset=['County'])
df1['County'] = df1['County'].astype(str)
df2 = df1.merge(df, how='left', on=['County'])
print(df2['County'].nunique())
```

```python
# Cross Product
df3 = pd.DataFrame(df2['IssuerId'].unique())
df4 = pd.DataFrame(df2['County'].unique())
print(df3)
merged = df3.merge(df4, how='cross')
#merged = merged.drop_duplicates()
merged
```
