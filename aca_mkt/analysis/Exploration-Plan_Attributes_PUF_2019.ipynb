```python
import pandas as pd
import numpy as np
```

```python
data_path =  "../data/data_2019"
df = pd.read_csv(data_path + 'Plan_Attributes_PUF_2019.csv', encoding='cp1252')
```

```python
print('There are', df['IssuerId'].nunique(), 'unique IssuerIds in the data.')
```

```python
print('There are', df['PlanMarketingName'].nunique(), 'unique IssuerMarketPlaceMarketingNames in the data.')
```

```python
df.groupby('SourceName').agg({'IssuerId': 'nunique', 'PlanMarketingName': 'nunique'})
```

```python
summary = pd.DataFrame()
summary['Column_Name'] = pd.Series(list(df.columns))
summary
```

```python
DEHB_list = list(df.filter(like='TEHB').columns)
MEHB_list = list(df.filter(like='MEHB').columns)
TEHB_list = list(df.filter(like='DEHB').columns)
```

```python
summary['Column_2'] = np.where(summary['Column_Name'].isin(TEHB_list), 1, 0 )
summary['Column_3'] = np.where(summary['Column_Name'].isin(MEHB_list), 2, 0 )
summary['Column_4'] = np.where(summary['Column_Name'].isin(DEHB_list), 3, 0 )
summary['Categories'] = summary['Column_2'] + summary['Column_3'] + summary['Column_4']
summary = summary[['Column_Name', 'Categories']]
summary
```

```python
lst = []
for i in list(df.columns):
    lst.append(df[i].isna().sum())
summary['Missing_Values'] = pd.Series(lst)
summary
```

```python
lst = []
for i in list(df.columns):
    lst.append(df[i].nunique())
summary['Unique_Values'] = pd.Series(lst)
summary
```

```python
summary.to_csv('plan_summary.csv')
```
