```python
import numpy as np
import pandas as pd
```

```python
columns = ['state_code','prov_nbr','city','state','zip','ssa']
result = pd.DataFrame(columns=columns)

with open('data/ESRD Provider PUF as of 01FEB2020.txt', 'rb') as f:
    contents = f.readlines()
    for row in contents[2:-20]:
        raw_data = row[0:2],row[2:6],row[186:206],row[206:208],row[208:213],row[259:262]
        raw_data = [str(i)[2:-1].replace(' ','') for i in raw_data]
        new_row = pd.DataFrame(data=[raw_data],columns=columns)
        result = result.append(new_row)

result = result.reset_index()[columns]

result.to_csv('data/ESRD_Provider_PUF.csv')
```

```python
patient_summary =  pd.read_csv("data/patient_summary.csv")
treatment = pd.read_csv("data/admit_treatment.csv")

print(patient_summary.columns,treatment.columns)
```
